# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# 🛠️ TOOL 3 (Bonus): Word Counter

def count_words(text: str) -> int:
    """Count the number of words in a piece of text."""
    try:
        return len(text.split())
    except Exception:
        return 0
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION

agent_logs = []  # keeps a running log of every call, used for the logging bonus below

def agent(query: str):
    query_lower = query.lower()

    try:
        if "calculate" in query_lower:
            # strip out the word "calculate" and just keep the math expression
            expression = query_lower.replace("calculate", "").strip()

            if not expression:
                response = {"type": "error", "result": "No expression found to calculate"}
            else:
                calc_result = calculator(expression)
                if calc_result == "Error in calculation":
                    response = {"type": "error", "result": calc_result}
                else:
                    response = {"type": "calculation", "result": calc_result}

        elif "keywords" in query_lower:
            response = {"type": "keywords", "result": extract_keywords(query)}

        elif "word count" in query_lower or "count words" in query_lower:
            # bonus tool, see below
            response = {"type": "word_count", "result": count_words(query)}

        else:
            response = {
                "type": "general",
                "result": f"I don't have a specific tool for that, here's your query back: \"{query}\""
            }

    except Exception as e:
        response = {"type": "error", "result": f"Something went wrong: {e}"}

    # log every call (query + what the agent decided) for the logging bonus
    agent_logs.append({"query": query, "response": response})

    return response
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate 100 / 0",          # tests the error handling
    "Give me the word count of this sentence please"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

## 🚀 Bonus: Logging

Every call to `agent()` gets appended to `agent_logs`, so we can review what happened after the fact.

````python
# check the logs
for entry in agent_logs:
    print(entry)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
